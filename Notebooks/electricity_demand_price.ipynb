```python
import pandas as pd
```

```python
#read the csv 

df_price = pd.read_csv("../Resources/electricity_price.csv")
df_price.head(10)
```

```python
df_price.describe()
```

```python
df_price = df_price.drop(["Series Key.1", "Series Key.2", "Series Key.3","ELEC.PRICE.US-ALL.M"], axis=1)
df_price.head(10)
```

```python
df_price=df_price.drop([0,1,2,3,4,5])
```

```python
df_price.head(10)
```

```python
df_price_final = df_price.sort_values(by=['Series Key'], ascending = True).reset_index(drop=True)
df_price_final.head(10)
```

```python
df_price_final=df_price_final.rename(columns={ "ELEC.PRICE.US-RES.M": "residential_price", "ELEC.PRICE.US-COM.M": "commercial_price", "ELEC.PRICE.US-IND.M":"industrial_price", "Series Key":"date_id"})
df_price_final.head()
```
